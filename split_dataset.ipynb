```python
df = pd.read_csv('data/e_commerce_shopper_behaviour_and_lifestyle.csv')
```

```python
import random

# 1. On isole l'ID et on prépare le reste
id_col = ["user_id", "last_purchase_date"]
# On récupère toutes les colonnes sauf l'ID
other_cols = [c for c in df.columns if c not in id_col]

# 2. Mélange des colonnes restantes (en convertissant en liste pour éviter le TypeError)
other_cols = list(other_cols)
random.seed(42)
random.shuffle(other_cols)

# 3. Calcul du point de coupe
split_index = len(other_cols) // 2
cols_part_1 = other_cols[:split_index]
cols_part_2 = other_cols[split_index:]

# 4. Création des deux DataFrames en incluant l'ID dans chaque sélection
# Syntaxe Pandas : df[[liste_de_colonnes]]
df_1 = df[id_col + cols_part_1].copy()
df_2 = df[id_col + cols_part_2].copy()

# Vérification
print(f"Colonnes DF 1 : {list(df_1.columns)}")
print(f"Colonnes DF 2 : {list(df_2.columns)}")
```
